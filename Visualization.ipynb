# Matplotlib

```python
Categories = df ['Category'].value_counts()
Categories
```

```python
plt.plot(Categories, 'r')  
# يرسم خط بياني للبيانات الموجودة في Categories واللون r يعني أحمر

plt.xticks(rotation=45)  
# يلف أسماء الفئات في المحور الأفقي بزاوية 45 درجة عشان تصير أوضح

plt.title('Orders by Category', fontsize=15)  
# يضيف عنوان للرسم ويحدد حجم الخط

plt.xlabel('Category')  
# يحدد اسم المحور الأفقي

plt.ylabel('Orders')  
# يحدد اسم المحور العمودي

plt.show()  
# يعرض الرسم البياني
```

```python
country_orders = df['Country'].value_counts()
# يحسب عدد التكرارات لكل دولة في عمود Country
# يعني يشوف كل دولة كم مرة تكررت في البيانات

country_orders
# يعرض نتيجة العد (عدد الطلبات لكل دولة)
```

```python
plt.bar(country_orders.index, country_orders.values , color = 'g')
# يرسم مخطط أعمدة (Bar Chart)
# index = أسماء الدول
# values = عدد الطلبات لكل دولة

plt.xticks(rotation=90)
# يلف أسماء الدول 90 درجة عشان تصير مرتبة وواضحة

plt.xlabel('Country')
# يحدد اسم المحور الأفقي (الدول)

plt.ylabel('Orders')
# يحدد اسم المحور العمودي (عدد الطلبات)

plt.grid(axis='y', color='r', alpha=0.35, linestyle=':')
# يضيف خطوط شبكة على المحور الرأسي فقط
# axis='y'  → يطبق الشبكة على محور Y
# color='g' → لون الخطوط الاحمز
# alpha=0.35 → درجة الشفافية
# linestyle=':' → الخط يكون منقط

plt.show()
# يعرض الرسم البياني
```
